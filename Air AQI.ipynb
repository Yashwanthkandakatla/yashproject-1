Considering a new dataframe named air

```python
air= pd.read_csv('city_day.csv')
```

Filling the NAN(Not a number) values with some values

```python
air = air.fillna(air.mean())
```

Converting the qualitative values into quantitative values
(Sequencing the cities)

```python
cities = air['City']
cityset = set(cities)
cl = list(cityset)
dictofcities = {cl[i] : i for i in range(0,len(cl))}
air['City'] = air['City'].map(dictofcities)
```

```python
air.head()
```

The AQI_Bucket is not required for training the model

```python
air.head()
```

```python
air.isnull().sum()
```

Removing the AQI_Bucket column

```python
air = air.drop('AQI_Bucket',axis=1)
```

```python
air = air.drop('Date',axis=1)
```

```python
air.head()
```

Considering the independent and dependent variables in order to train the model

```python
independent = air[['City','PM2.5','PM10','NO','NO2','NOx','NH3','CO','SO2','O3','Benzene','Toluene','Xylene','AQI']]

dependent = air['AQI']
```

Splitting the data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(independent,dependent, test_size=0.33, random_state=42)
```

Training the data using RandomForest algorithm

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=2,random_state=0)
regr.fit(X_train,y_train)
```

Predicting the output using RandomForest Algorithm

```python
predicted = regr.predict(X_test)
print(regr.predict(X_test))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,predicted)
```
